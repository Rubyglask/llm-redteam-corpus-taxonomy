## Load via Parquet directly

```python
import pandas as pd
df = pd.read_parquet(
    "hf://datasets/rubyglask/llm-redteam-corpus-taxonomy-core/canonical.parquet"
)
df.groupby(["source_id", "language"]).size()
```

## DuckDB analytics

```python
import duckdb
con = duckdb.connect("canonical.duckdb")
con.sql("""
  SELECT language, source_id, count(*) AS n
  FROM canonical
  GROUP BY 1, 2 ORDER BY n DESC
""").df()
```
